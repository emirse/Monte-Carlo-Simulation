```python
import random
import numpy as np
import pandas as pd
import operator
```

```python
df = pd.read_csv('simulasyon.csv', encoding='unicode_escape', skipinitialspace=True)
dizi2 = []
```

```python
def tarih():
    return df['Ay']
```

```python
def hesaplaFrekans(data,value):
    miktar = dict(data[value].value_counts())
    sozlukListele = sorted(miktar.items(), key=operator.itemgetter(0))
    miktar = pd.DataFrame(sozlukListele)

    frekansDf = pd.DataFrame(miktar)
    frekansDfToplam = frekansDf[1].sum()
    frekansAdet = frekansDf[1]
    frekansMiktar = frekansDf[0]
    if value==1:
        print(miktar)
    else:

        return frekansAdet, frekansDfToplam, frekansMiktar
```

```python
def sayiUret():
    randomDizi = []
    gun = tarih()

    for i in range(12):
        randomDizi.append(random.random())

    return randomDizi
```

```python
def hesaplaOlasilik():
    bululanDegerler = hesaplaFrekans(df, "Miktar")

    for i in bululanDegerler[0]:
        print(f"Frekansın olasılık {i} girdisi sonucu:",
              i / bululanDegerler[1])

        dizi2.append(i / bululanDegerler[1])
    return dizi2, bululanDegerler
```

```python
olasilikHesap = hesaplaOlasilik()
```

```python
def kumulatif(deger):
    sums = []
    sums.append(0)
    for i in range(len(deger)):
        sums.append(deger[i] + sums[i])
    for j in sums:
        print("Kümülatif olasılık:", j)
    return sums
```

```python
kumDeger = kumulatif(olasilikHesap[0])
```

```python
def miktarAyari(kumDeger, df, olasilikHesap):
    oncekiMiktar = olasilikHesap[1][2]
    gun = tarih()
    yeniDizi = []
    for i in range(len(kumDeger) - 1):
        for j in range(len(df)):
            if kumDeger[i] < df[j] < kumDeger[i + 1]:
                yeniDizi.append([gun[j], oncekiMiktar[i]])
                print(f"{gun[j]} ayının miktarı :{oncekiMiktar[i]}")
    yeniDf = pd.DataFrame(yeniDizi)
    return yeniDf
```

```python
a = int(input("Kaç yıllık veri hesaplansın : \n"))
for i in range(a):
    miktarAyarlandi = miktarAyari(kumDeger, sayiUret(), olasilikHesap)
print(f"Geçen yıl : {i}")
hesaplaFrekans(miktarAyarlandi, 1)
```
